```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
Z = X**2 + Y**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
pokemon_data = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
pokemon_data.head()
```

```python
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = pokemon_data[features]
```

```python
pearson_corr = data.corr(method='pearson')
```

```python
spearman_corr = data.corr(method='spearman')
```

```python
def plot_heatmap(corr_matrix, title):
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True,
                linewidths=0.5, linecolor='black')
    plt.title(title)
    plt.show()
```

```python
plot_heatmap(pearson_corr, 'Pearson Correlation Heatmap')
```

```python
plot_heatmap(spearman_corr, 'Spearman Rank Correlation Heatmap')
```
