```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv('Au_nanoparticle_dataset.csv')
```

```python
data_set.head()
```

```python
columns_to_drop = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
```

```python
filtered_data_set = data_set.drop(columns=columns_to_drop)
```

```python
filtered_data_set.head(20)
```

```python
statistics = {}
for column in columns_to_drop:
    statistics[column] = {
        'mean': data_set[column].mean(),
        'std_dev': data_set[column].std(),
        '25%': data_set[column].quantile(0.25),
        '50%': data_set[column].quantile(0.50),
        '75%': data_set[column].quantile(0.75)
    }
```

```python
stats_df = pd.DataFrame(statistics)
```

```python
stats_df = stats_df.T
```

```python
print(stats_df)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

for i, column in enumerate(columns_to_drop):
    axes[i].hist(data_set[column], bins=30, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Histogram of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()

plt.show()
```

```python
filterd_data = data_set[columns_to_drop]
```

```python
pairplot = sns.pairplot(filterd_data)

pairplot.fig.suptitle("Pairplot of N_total, N_bulk, N_surface, and R_avg", y=1.02)
plt.show()
```

```python
g = sns.PairGrid(filterd_data)

g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)

g.fig.suptitle("Pairplot of N_total, N_bulk, N_surface, and R_avg", y=1.02)
plt.show()
```
